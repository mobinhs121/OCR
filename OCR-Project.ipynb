```python
import pytesseract
from PIL import Image
import cv2
import numpy as np
import os
import pandas as pd
import joblib
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, LSTM, Dense, Embedding
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics.pairwise import cosine_similarity

# Function to preprocess and extract text from an image
def preprocess_and_extract_text(image_path):
    try:
        # Open the image
        img = Image.open(image_path)

        # Convert the image to grayscale
        img_gray = img.convert('L')

        # Convert PIL image to NumPy array
        img_np = np.array(img_gray)

        # Perform thresholding to binarize the image (adjust threshold value as needed)
        _, img_bin = cv2.threshold(img_np, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Perform OCR using Tesseract on the preprocessed image
        text = pytesseract.image_to_string(Image.fromarray(img_bin), lang='eng+ara+fas', config='--psm 6')

        return text
    except Exception as e:
        return str(e)

# Define paths to your datasets
data_path = r'/content/data'
data2_path = r'/content/data2'
scan_path = r'/content/scan'
scan_split_path = r'/content/scan splited'

# Load data from the standard-ocr-dataset (data folder)
standard_ocr_data_path = os.path.join(data_path, 'training_data')  # Adjust as needed
image_text_pairs = []

for folder_name in os.listdir(standard_ocr_data_path):
    folder_path = os.path.join(standard_ocr_data_path, folder_name)
    if os.path.isdir(folder_path):
        for image_name in os.listdir(folder_path):
            image_path = os.path.join(folder_path, image_name)
            text = folder_name  # You can adjust this based on your data structure
            image_text_pairs.append({'image_path': image_path, 'text': text})

# Load data from scan folder
scan_data_path = os.path.join(scan_path, 'scan')
for subfolder_name in os.listdir(scan_data_path):
    subfolder_path = os.path.join(scan_data_path, subfolder_name)
    if os.path.isdir(subfolder_path):
        for image_name in os.listdir(subfolder_path):
            image_path = os.path.join(subfolder_path, image_name)
            text = subfolder_name  # You can adjust this based on your data structure
            image_text_pairs.append({'image_path': image_path, 'text': text})

# Split data into training and testing sets
df = pd.DataFrame(image_text_pairs, columns=['image_path', 'text'])
X_train, X_test, y_train, y_test = train_test_split(df['image_path'], df['text'], test_size=0.2, random_state=42)

# Tokenize text data
tokenizer = Tokenizer()
tokenizer.fit_on_texts(y_train)
vocab_size = len(tokenizer.word_index) + 1
max_sequence_length = 100

X_train_text = tokenizer.texts_to_sequences(y_train)
X_test_text = tokenizer.texts_to_sequences(y_test)

X_train_text = pad_sequences(X_train_text, maxlen=max_sequence_length, padding='post')
X_test_text = pad_sequences(X_test_text, maxlen=max_sequence_length, padding='post')

# Encode text labels
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)

# Define the deep learning model
embedding_dim = 64
image_input = keras.layers.Input(shape=(1,))
text_input = keras.layers.Input(shape=(max_sequence_length,))

# Image branch
image_embedding = keras.layers.Embedding(len(X_train), embedding_dim)(image_input)
image_flatten = keras.layers.Flatten()(image_embedding)

# Text branch
text_embedding = keras.layers.Embedding(vocab_size, embedding_dim)(text_input)
text_lstm = keras.layers.LSTM(128)(text_embedding)

# Merge branches
merged = keras.layers.concatenate([image_flatten, text_lstm])
dense1 = keras.layers.Dense(128, activation='relu')(merged)
output = keras.layers.Dense(len(encoder.classes_), activation='softmax')(dense1)

model = Model(inputs=[image_input, text_input], outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
# Train the model (continued)
model.fit([np.arange(len(X_train)), X_train_text], y_train_encoded, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
accuracy = model.evaluate([np.arange(len(X_test)), X_test_text], y_test_encoded)[1]
print(f"Accuracy: {accuracy * 100:.2f}%")

# Save the model
model.save('image_text_matching_model.h5')
print("Model saved as 'image_text_matching_model.h5'")
```

```python
import pytesseract
from PIL import Image
import cv2
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, Flatten, Dense, Embedding, LSTM, Lambda
)
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Function to preprocess and extract text from an image using Tesseract
def tesseract_extract_text(image_path):
    try:
        # Open the image
        img = Image.open(image_path)

        # Convert the image to grayscale
        img_gray = img.convert('L')

        # Convert PIL image to NumPy array
        img_np = np.array(img_gray)

        # Perform thresholding to binarize the image (adjust threshold value as needed)
        _, img_bin = cv2.threshold(img_np, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Perform OCR using Tesseract on the preprocessed image
        text = pytesseract.image_to_string(Image.fromarray(img_bin), lang='eng+ara+fas', config='--psm 6')

        return text
    except Exception as e:
        return str(e)

# Define paths to your datasets (make sure they are correct)
data_path = '/content/data'
data2_path = '/content/data2'
scan_path = '/content/scan'
scan_split_path = '/content/scan splited'

# Create a dataset of image-text pairs
image_text_pairs = []

# Function to load images and extract text using Tesseract
def load_and_process_data(data_dir):
    for subdir in os.listdir(data_dir):
        subdir_path = os.path.join(data_dir, subdir)
        if os.path.isdir(subdir_path):
            for filename in os.listdir(subdir_path):
                if filename.endswith(".jpg"):
                    image_path = os.path.join(subdir_path, filename)
                    tesseract_text = tesseract_extract_text(image_path)
                    image_text_pairs.append({'image_path': image_path, 'tesseract_text': tesseract_text})

# Load data from various sources
load_and_process_data(data_path)
load_and_process_data(data2_path)
load_and_process_data(scan_path)
load_and_process_data(scan_split_path)

# Shuffle the dataset
random.shuffle(image_text_pairs)

# Split the dataset into training and testing sets
train_size = int(0.8 * len(image_text_pairs))
train_pairs = image_text_pairs[:train_size]
test_pairs = image_text_pairs[train_size:]

# Preprocessing parameters
MAX_TEXT_LENGTH = 100  # Max length of text sequences
IMAGE_WIDTH = 128    # Image width after preprocessing
IMAGE_HEIGHT = 128   # Image height after preprocessing
IMAGE_CHANNELS = 3   # Number of image channels (RGB)

# Debugging Tip 1: Print the length of your datasets
print("Total pairs:", len(image_text_pairs))
print("Training pairs:", len(train_pairs))
print("Testing pairs:", len(test_pairs))

# Implement advanced image preprocessing
def preprocess_image(image_path):
    # Load and preprocess the image
    img = Image.open(image_path)

    # Resize and normalize the image
    img = img.resize((IMAGE_WIDTH, IMAGE_HEIGHT))
    img = np.array(img) / 255.0

    # Apply additional image preprocessing steps if needed (e.g., denoising, contrast enhancement)

    return img

# Implement text processing
def preprocess_text(text):
    # Tokenization
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts([text])
    text_sequence = tokenizer.texts_to_sequences([text])

    # Padding
    text_padded = pad_sequences(text_sequence, maxlen=MAX_TEXT_LENGTH, padding='post')

    return text_padded

# Implement model architecture for text-image comparison
def build_model():
    # Image input
    image_input = Input(shape=(IMAGE_WIDTH, IMAGE_HEIGHT, IMAGE_CHANNELS))

    # Implement convolutional layers for image processing
    x = Conv2D(32, (3, 3), activation='relu')(image_input)
    x = MaxPooling2D((2, 2))(x)
    x = Flatten()(x)
    x = Dense(64, activation='relu')(x)

    # Text input
    text_input = Input(shape=(MAX_TEXT_LENGTH,))

    # Implement text processing layers (e.g., word embeddings)
    embedding_layer = Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=32)(text_input)
    lstm_layer = LSTM(64)(embedding_layer)

    # Combine image and text features
    combined_features = tf.concat([x, lstm_layer], axis=-1)

    # Implement a similarity layer (e.g., cosine similarity)
    similarity_score = Lambda(lambda x: tf.reduce_sum(x[0] * x[1], axis=1) / (tf.norm(x[0], axis=1) * tf.norm(x[1], axis=1)))(
        [combined_features, combined_features])

    # Build the model
    model = Model(inputs=[image_input, text_input], outputs=similarity_score)

    return model

# Compile the model with appropriate loss function and optimizer
model = build_model()
model.compile(loss='cosine_similarity', optimizer='adam')

# Debugging Tip 2: Print the model summary to check the architecture
model.summary()

# Define a function to generate training batches with a specified number of batches
def generate_training_batches(batch_size, num_batches):
    for _ in range(num_batches):
        random.shuffle(train_pairs)
        for i in range(0, len(train_pairs), batch_size):
            batch = train_pairs[i:i + batch_size]
            images = []
            texts = []
            labels = []
            for item in batch:
                image_path = item['image_path']
                tesseract_text = item['tesseract_text']

                # Preprocess the image
                image = preprocess_image(image_path)

                # Preprocess the Tesseract text
                text = preprocess_text(tesseract_text)

                # Create positive and negative pairs for training
                positive_image = image
                positive_text = text
                for j in range(batch_size):
                    negative_image = random.choice(train_pairs)['image_path']
                    negative_image = preprocess_image(negative_image)
                    negative_text = random.choice(train_pairs)['tesseract_text']
                    negative_text = preprocess_text(negative_text)
                    images.append([positive_image, negative_image])
                    texts.append([positive_text, negative_text])
                    labels.append(1)  # Positive pair
                    labels.append(0)  # Negative pair
            yield [np.array(images), np.array(texts)], np.array(labels)

# Example: Generate 10 batches of training data
num_training_batches = 10
batch_size = 32

for batch_data, batch_labels in generate_training_batches(batch_size, num_training_batches):
    # Your training code here
    pass  # Remove this pass statement and add your training code

# Save the trained model
model.save('/content/ocr_model.h5')
```

```python
import pytesseract
from PIL import Image
import cv2
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, Flatten, Dense, Embedding, LSTM, Lambda
)
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Function to preprocess and extract text from an image using Tesseract
def tesseract_extract_text(image_path):
    try:
        # Open the image
        img = Image.open(image_path)

        # Convert the image to grayscale
        img_gray = img.convert('L')

        # Convert PIL image to NumPy array
        img_np = np.array(img_gray)

        # Perform thresholding to binarize the image (adjust threshold value as needed)
        _, img_bin = cv2.threshold(img_np, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Perform OCR using Tesseract on the preprocessed image
        text = pytesseract.image_to_string(Image.fromarray(img_bin), lang='eng+ara+fas', config='--psm 6')

        return text
    except Exception as e:
        return str(e)

# Define paths to your datasets (make sure they are correct)
data_path = '/content/data'
data2_path = '/content/data2'
scan_path = '/content/scan'
scan_split_path = '/content/scan splited'

# Create a dataset of image-text pairs
image_text_pairs = []

# Function to load images and extract text using Tesseract
def load_and_process_data(data_dir):
    for subdir in os.listdir(data_dir):
        subdir_path = os.path.join(data_dir, subdir)
        if os.path.isdir(subdir_path):
            for filename in os.listdir(subdir_path):
                if filename.endswith(".jpg"):
                    image_path = os.path.join(subdir_path, filename)
                    tesseract_text = tesseract_extract_text(image_path)
                    image_text_pairs.append({'image_path': image_path, 'tesseract_text': tesseract_text})

# Load data from various sources
load_and_process_data(data_path)
load_and_process_data(data2_path)
load_and_process_data(scan_path)
load_and_process_data(scan_split_path)

# Shuffle the dataset
random.shuffle(image_text_pairs)

# Split the dataset into training and testing sets
train_size = int(0.8 * len(image_text_pairs))
train_pairs = image_text_pairs[:train_size]
test_pairs = image_text_pairs[train_size:]

# Preprocessing parameters
MAX_TEXT_LENGTH = 100  # Max length of text sequences
IMAGE_WIDTH = 128    # Image width after preprocessing
IMAGE_HEIGHT = 128   # Image height after preprocessing
IMAGE_CHANNELS = 3   # Number of image channels (RGB)

# Debugging Tip 1: Print the length of your datasets
print("Total pairs:", len(image_text_pairs))
print("Training pairs:", len(train_pairs))
print("Testing pairs:", len(test_pairs))

# Implement advanced image preprocessing
def preprocess_image(image_path):
    # Load and preprocess the image
    img = Image.open(image_path)

    # Resize and normalize the image
    img = img.resize((IMAGE_WIDTH, IMAGE_HEIGHT))
    img = np.array(img) / 255.0

    # Apply additional image preprocessing steps if needed (e.g., denoising, contrast enhancement)

    return img

# Implement text processing
def preprocess_text(text):
    # Tokenization
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts([text])
    text_sequence = tokenizer.texts_to_sequences([text])

    # Padding
    text_padded = pad_sequences(text_sequence, maxlen=MAX_TEXT_LENGTH, padding='post')

    return text_padded

# Implement model architecture for text-image comparison
def build_model():
    # Image input
    image_input = Input(shape=(IMAGE_WIDTH, IMAGE_HEIGHT, IMAGE_CHANNELS))

    # Implement convolutional layers for image processing
    x = Conv2D(32, (3, 3), activation='relu')(image_input)
    x = MaxPooling2D((2, 2))(x)
    x = Flatten()(x)
    x = Dense(64, activation='relu')(x)

    # Text input
    text_input = Input(shape=(MAX_TEXT_LENGTH,))

    # Implement text processing layers (e.g., word embeddings)
    embedding_layer = Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=32)(text_input)
    lstm_layer = LSTM(64)(embedding_layer)

    # Combine image and text features
    combined_features = tf.concat([x, lstm_layer], axis=-1)

    # Implement a similarity layer (e.g., cosine similarity)
    similarity_score = Lambda(lambda x: tf.reduce_sum(x[0] * x[1], axis=1) / (tf.norm(x[0], axis=1) * tf.norm(x[1], axis=1)))(
        [combined_features, combined_features])

    # Build the model
    model = Model(inputs=[image_input, text_input], outputs=similarity_score)

    return model

# Compile the model with appropriate loss function and optimizer
model = build_model()
model.compile(loss='cosine_similarity', optimizer='adam')

# Debugging Tip 2: Print the model summary to check the architecture
model.summary()

# Train the model using the training dataset
# Define a function to generate training batches
# Define a function to generate training batches with a specified number of batches
def generate_training_batches(batch_size, num_batches):
    for _ in range(num_batches):
        random.shuffle(train_pairs)
        for i in range(0, len(train_pairs), batch_size):
            batch = train_pairs[i:i + batch_size]
            images = []
            texts = []
            labels = []
            for item in batch:
                image_path = item['image_path']
                tesseract_text = item['tesseract_text']

                # Preprocess the image
                image = preprocess_image(image_path)

                # Preprocess the Tesseract text
                text = preprocess_text(tesseract_text)

                # Create positive and negative pairs for training
                positive_image = image
                positive_text = text
                for j in range(batch_size):
                    negative_image = random.choice(train_pairs)['image_path']
                    negative_image = preprocess_image(negative_image)
                    negative_text = random.choice(train_pairs)['tesseract_text']
                    negative_text = preprocess_text(negative_text)
                    images.append([positive_image, negative_image])
                    texts.append([positive_text, negative_text])
                    labels.append(1)  # Positive pair
                    labels.append(0)  # Negative pair
            yield [np.array(images), np.array(texts)], np.array(labels)

# Example: Generate 10 batches of training data
num_training_batches = 10
batch_size = 32

for batch_data, batch_labels in generate_training_batches(batch_size, num_training_batches):
    # Your training code here
   # Remove this pass statement and add your training code

# Note: Adjust 'num_training_batches' as needed based on the number of batches you want to generate.


# Train the model
    batch_size = 32
    epochs = 10

    train_generator = generate_training_batches(batch_size)
    steps_per_epoch = len(train_pairs) // batch_size

# Debugging Tip 3: Print the shape of input data to verify it matches the model's input shape
    sample_batch = next(train_generator)
    print("Input shapes - Images:", sample_batch[0][0].shape, "Texts:", sample_batch[0][1].shape)
    print("Label shape:", sample_batch[1].shape)

    model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=epochs,
    verbose=1,
)

# Evaluate the model using the testing dataset
# Define a function to generate testing batches
    def generate_testing_batches(batch_size):
        for i in range(0, len(test_pairs), batch_size):
            batch = test_pairs[i:i + batch_size]
            images = []
            texts = []
            labels = []
            for item in batch:
                image_path = item['image_path']
                tesseract_text = item['tesseract_text']

            # Preprocess the image
                image = preprocess_image(image_path)

            # Preprocess the Tesseract text
                text = preprocess_text(tesseract_text)

            # Create positive and negative pairs for testing
                positive_image = image
                positive_text = text
                for j in range(batch_size):
                    negative_image = random.choice(test_pairs)['image_path']
                    negative_image = preprocess_image(negative_image)
                    negative_text = random.choice(test_pairs)['tesseract_text']
                    negative_text = preprocess_text(negative_text)
                    images.append([positive_image, negative_image])
                    texts.append([positive_text, negative_text])
                    labels.append(1)  # Positive pair
                    labels.append(0)  # Negative pair
            yield [np.array(images), np.array(texts)], np.array(labels)

# Evaluate the model
    test_batch_size = 32
    test_generator = generate_testing_batches(test_batch_size)
    test_steps = len(test_pairs) // test_batch_size

# Debugging Tip 4: Print the shape of input data to verify it matches the model's input shape for testing
    sample_test_batch = next(test_generator)
    print("Testing Input shapes - Images:", sample_test_batch[0][0].shape, "Texts:", sample_test_batch[0][1].shape)
    print("Testing Label shape:", sample_test_batch[1].shape)

    evaluation = model.evaluate(
    test_generator,
    steps=test_steps,
    verbose=1,
)

# Print evaluation results
    print("Evaluation Loss:", evaluation)
# Save the trained model
model.save('/content/ocs_model.h5')
```
